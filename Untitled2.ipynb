```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import tessreduce as tr
```

```python
ra = 189.1385817
dec = 11.2316535
#ra = 64.526125
#dec = -63.61506944
tpf = tr.Get_TESS(ra,dec,Sector=23)
```

```python
red = tr.Quick_reduce(tpf,diff_lc=True)
```

```python
bkg = red['bkg']
raw = tpf.flux.value
```

```python

circle = plt.Circle((45, 45), 5, color='r', fill=False,lw=2)

plt.figure(figsize=(9,2.8))
plt.subplot(141)
plt.title('Raw image')
plt.imshow(raw[700],origin='lower',vmin=np.percentile(raw[700],16),vmax=np.percentile(raw[700],80))
plt.axis('off')

plt.subplot(142)
plt.title('Background')
plt.imshow(bkg[700],origin='lower',vmin=np.percentile(raw[700],16),vmax=np.percentile(raw[700],80))
plt.axis('off')

plt.subplot(143)
plt.title('Background subtracted')
sub = raw[700] - bkg[700]
plt.imshow(sub,origin='lower',vmin=np.percentile(sub,16),vmax=np.percentile(sub,95))
plt.axis('off')

plt.subplot(144)
plt.title('Difference image')
sub = red['flux'][700] - red['flux'][100]
plt.imshow(sub,origin='lower',vmin=np.percentile(sub,16),vmax=np.percentile(sub,99))
ax = plt.gca()
ax.add_patch(circle)
plt.axis('off')

plt.tight_layout()

plt.savefig('tess_reduction.pdf')
```
